```python
import matplotlib.pyplot as plt
from keras.models import load_model
from PIL import Image
import numpy as np

model = load_model('mnistEpochs30CNN.h5')

def testList(name):
    digit = Image.open("/home/chen/LiChiChang/2020SpringThesis/dataset/mnist/test/"+str(name)+".jpg")
    digit = digit.resize(size=(28,28))
    digit = digit.convert('L')
    digit = np.array(digit)
    digit = digit.reshape(1,28,28,1)
    for i in range(28):
        #string = ''
        for j in range(28):
            digit[0,i,j,0] = 255 - digit[0,i,j,0]
            if digit[0,i,j,0]<120:
                digit[0,i,j,0] = 0
            #string = string + str(digit[0,i,j,0]) + '\t'
        #print(string)
    digit = digit/255.0

    plt.imshow(digit.reshape(28,28), cmap=plt.cm.get_cmap('binary'))
    plt.show()

    print(model.predict(np.array(digit)))
    print(model.predict_classes(np.array(digit)))

for i in range(20):
    testList(i+1)
```
